# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Create Classifier

```python
face_classifier = cv2.CascadeClassifier('anaconda3\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml')
```

# Open Image

```python
image = cv2.imread('Images/baby.jpg')
fix_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

# Detect Faces

```python
faces = face_classifier.detectMultiScale(image, 1.3, 5)
```

# No Faces Detected

```python
if faces is():
    print('No Faces found')
```

# Create a function to detect face

```python
def detect_face(fix_img):
    face_rects = face_classifier.detectMultiScale(fix_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(fix_img,
                     (x,y),
                     (x+w, y+h),
                     (255,0,0),
                     10)
    return fix_img
```

# Result

```python
result = detect_face(fix_img)
plt.imshow(result)
```
